```python
import pandas as pd
import folium
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('/home/breno/Desktop/Estudo_COVID-19/time_series_covid_19_confirmed.csv')
#Para a exibicao do mapa funcionar. Voce deve copiar o caminho do arquivo CSV que eu forneci a voces
#e colar aqui em cima.
```

```python
data = data[['Lat','Long']]
```

```python
data.head()
```

```python
from IPython.display import display
centro = (data.iloc[0]['Lat'], data.iloc[0]['Long'])
mapa = folium.Map(location=centro, zoom_start=1.3)
for i in range(len(data)):
    linha = data.iloc[i]
    ponto = (linha['Lat'], linha['Long'])
    marcador = folium.Marker(ponto)
    marcador.add_to(mapa)
display(mapa)
mapa.save('mapa.html') #Isso vai salvar um HTML do mapa que vc pode tirar ScreenShot
#    Daqui pra baixo sao so testes.
#plt.savefig(display(mapa))
#savefig(display(mapa), bbox_inches='tight')
```
